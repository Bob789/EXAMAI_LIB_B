```python
import numpy as np
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from CSV files and display their information
# we handel the missing values and duplicates in the dataframes in each section
print('-' * 20, 'df_items', '-' * 20)
df_items = pd.read_csv("items.csv")
display(df_items.info())
display(df_items.describe())
display(df_items.head())

print('-' * 20, 'df_order_items', '-' * 20)
df_order_items = pd.read_csv("order_item.csv")
display(df_order_items.info())
display(df_order_items.describe())
display(df_order_items.head())

print('-' * 20, 'df_orders', '-' * 20)
df_orders = pd.read_csv("orders.csv")
display(df_orders.info())
display(df_orders.describe())
display(df_orders.head())

print('-' * 20, 'df_customers', '-' * 20)
df_customers = pd.read_csv("customers.csv")
display(df_customers.info())
display(df_customers.describe())
display(df_customers.head())
```

```python
# Level 1: Dataframe of all rows with item_name duplicate
df_dup = df_items[
    df_items.groupby('item_name')['id'].transform('count') > 1
].copy()
display('# Level 1: Dataframe of all rows with emails duplicate', df_dup)
```

```python
# Level 2: Adding column 'filled_count' to df which counts the number of non-null values in each row
df_dup['filled_count'] = df_dup.notnull().sum(axis=1)
display("# Level 2: Adding column 'filled_count' to df which counts the number of non-null values in each row",df_dup)
print(" Total rows in df_dup: ", len(df_dup))
```

```python
# Level 3: Find the most complete row for each item_name
df_keep = df_dup.loc[
    df_dup.groupby('item_name')['filled_count'].idxmax()
].copy()
df_keep = df_keep.drop(columns=['filled_count'])
display("# Level 3: Find the best rows for each email", df_keep)
print("numbers of rows in df_keep: ", len(df_keep))
```

```python
# level 3.1 keep the id drop so we can search this id for replacement in other tables the correct id 
idx_drop = df_dup.index.difference(df_keep.index)
display("# level 3.1 keep the id drop so we can search this id foe replacement in other tables the correct id ",idx_drop)
print("numbers rows to delete ",len(idx_drop))
```

```python
# Level 3.2: DataFrame of all other duplicate rows to drop for replacement
df_drop = df_dup.loc[idx_drop].copy()
df_drop = df_drop.drop(columns=['filled_count'])
display("Level 3.2: DataFrame of rows to drop", df_drop)
print("numbers of rows in df_drop: ", len(df_drop))
```

```python
# Level 4: Choose the rows with no duplicates as unique
df_unique = df_items[
    df_items.groupby('item_name')['id'].transform('count') == 1
].copy()
display("# Level 4: Choose the rows with no duplicates as unique", df_unique)
print("numbers of rows in df_unique: ", len(df_unique))

print("The numbers rows that we should get ofter the replacement is ", len(df_unique) + len(df_keep))
```

```python
# Level 5: merge the two dataframes
df_final = pd.concat([df_unique, df_keep], ignore_index=True)
display("# Level 5: merge the two dataframes",df_final)
print("numbers of rows in df_final: ", len(df_final))
```

```python
# Step 1: Prepare DataFrames for ID replacement ***************************
#   - df_drop: rows to remove, with 'id' renamed to old_id
#   - df_keep: rows to keep, with 'id' renamed to new_id

df_drop_map = df_drop[['item_name', 'id']].rename(columns={'id': 'old_id'})
df_keep_map = df_keep[['item_name', 'id']].rename(columns={'id': 'new_id'})

# Step 2: Merge on 'email' to build an old_id  new_id mapping table
df_id_map = pd.merge(
    df_drop_map,
    df_keep_map,
    on='item_name',
    how='left'
)

# df_id_map now contains columns 'item_name', 'old_id', and 'new_id'
display("# Step 2: Merge on 'item_name' to build an old_id  new_id mapping table",df_id_map)

# Step 3: Convert the mapping DataFrame to a dict for use with replace()
id_map = dict(zip(df_id_map['old_id'], df_id_map['new_id']))

# Display the mapping dictionary for verification
print("# Step 3: Convert the mapping DataFrame to a dict for use with replace()", id_map)

#---------------- checking the rows before changing ------------------
mask = df_order_items['item_id'].isin(df_drop_map['old_id'])
print("numbers of rows in df_order_items that we will change ", len(df_order_items[mask]))

# create DataFrame with the original index of the rows to be updated
rows_original = df_order_items[mask].copy()
display("rows to be updated: ", rows_original)

original_df_order_items = df_order_items['item_id'].copy()

#---------------- checking the mapping ------------------

# Step 4: Apply the replacement to df_order_items
df_order_items['item_id'] = (
    df_order_items['item_id']
    .replace(id_map)
)

# Verify the replacement
changed_rows = (original_df_order_items != df_order_items['item_id']).sum()
print(f"The number of changed rows whose affected: {changed_rows}")
```

```python

# Step 5: Drop rows with missing at least 3 values in df_customers
print("numbers of rows in df_customers before drop ", len(df_customers))
df_customers_cleaned = df_customers.dropna(thresh=len(df_customers.columns) - 2)
print("numbers of rows in df_customers_cleaned after drop ", len(df_customers_cleaned))

# Remove customers that are not in df_customers_cleaned
df_order_items_cleaned = df_order_items[
    df_order_items['id'].isin(df_customers_cleaned['id'])
].copy()





```

```python
# 2.1 Average item price
average_price = df_items['item_price'].mean()
print("Average item price:", average_price)
```

```python
# 2.2 The customer who purchased the most items

# Merge order items with orders to get customer_id for each item
merged = df_order_items_cleaned.merge(df_orders, on='order_id')

display("merged data frame", merged)

# Group by customer_id and calculate the total quantity each customer purchased
total_quantity_per_customer = merged.groupby('customer_id')['quantity'].sum()

display("total quantity per customer", total_quantity_per_customer)

# Find the customer_id with the highest total quantity
top_customer_id = total_quantity_per_customer.idxmax()
top_customer_quantity = total_quantity_per_customer.max()

# Merge orders with customers to get full customer info
merged_orders_customers = df_orders.merge(df_customers, left_on='customer_id', right_on='id')

# Extract the row with the top customer ID
top_customer_details = merged_orders_customers[merged_orders_customers['customer_id'] == top_customer_id]

# Print result..
print("Top customer ID:", top_customer_id)
print("Total quantity purchased:", top_customer_quantity)
print("Customer details:")
print(top_customer_details[['first_name', 'last_name', 'email', 'delivery_address']])
```

```python
# 2.1 Graph 

# --- Filter out customers with missing gender values ---
df_customers_cleaned = df_customers_cleaned.dropna(subset=['gender'])

# --- Count customers by gender 
gender_counts = df_customers_cleaned['gender'].value_counts()
print("Customers by gender:")
print(gender_counts)
gender_counts.plot(kind='bar', title='Customers by Gender')
plt.ylabel('Number of Customers')
plt.xlabel('Gender')
plt.tight_layout()
plt.show()
print("\n")

# --- Count customers by nationality 
nationality_counts = df_customers_cleaned['nationallity'].value_counts()
print("Customers by nationality:")
print(nationality_counts)
nationality_counts.plot(kind='bar', title='Customers by Nationality')
plt.ylabel('Number of Customers')
plt.xlabel('Nationality')
plt.tight_layout()
plt.show()



```

```python
# 2.3 Add total_price column to df_order_items_cleaned

# Display dataframe info before processing
display("df_order_items_cleaned before processing:", df_order_items_cleaned.info())

# Remove rows where item_id is missing (cannot calculate price without it)
df_order_items_cleaned = df_order_items_cleaned.dropna(subset=['item_id'])
display("df_order_items_cleaned after removing rows with missing item_id:", df_order_items_cleaned.info())

# Drop unnecessary or duplicate columns that may cause conflicts during merge
df_order_items_cleaned = df_order_items_cleaned.drop(
    columns=['id', 'item_price', 'price', 'item_price_x', 'item_price_y', 'price_x', 'price_y', 'id_x', 'id_y'],
    errors='ignore'
)

display("df_order_items_cleaned after dropping unnecessary columns:", df_order_items_cleaned.info())

# Merge with df_final to get item_price for each item_id
df_order_items_cleaned = df_order_items_cleaned.merge(
    df_final[['id', 'item_price']],
    left_on='item_id',
    right_on='id',
    how='left'
)

# Calculate the total price per item (quantity * item_price)
df_order_items_cleaned['total_price'] = (
    df_order_items_cleaned['quantity'] * df_order_items_cleaned['item_price']
)

# Drop the extra 'id' column from the merge result
df_order_items_cleaned = df_order_items_cleaned.drop(columns=['id'], errors='ignore')

# Display the final dataframe after adding total_price
display("df_order_items_cleaned after calculating total_price:", df_order_items_cleaned)


```

```python
# 2.4 Do group by order_id and sum the total_price per order
order_total_prices = df_order_items_cleaned.groupby('order_id')['total_price'].sum().reset_index()

# Display the result
display(order_total_prices)
```

```python
# 2.5 

# Step 1: Create total_price column in df_order_items_cleaned
order_summary = df_order_items_cleaned.groupby('order_id')['total_price'].sum().reset_index()

# Step 2: Most expensive order
most_expensive = order_summary.loc[order_summary['total_price'].idxmax()]

# Step 3: Cheapest order
cheapest = order_summary.loc[order_summary['total_price'].idxmin()]

# Step 4: Mean order price
average_total = order_summary['total_price'].mean()

# Step 5: Display results
print("Most expensive order:")
display(most_expensive.to_frame().T)

print("Cheapest order:")
display(cheapest.to_frame().T)

print("Average total price of orders:", average_total)
```

```python
# 2.5 Graph total sales per month

df_merged = df_order_items_cleaned.merge(df_orders[['order_id', 'order_date']], on='order_id')
df_merged['order_date'] = pd.to_datetime(df_merged['order_date'])
display(df_merged)

df_merged['year_month'] = df_merged['order_date'].dt.to_period('M')

sales_per_month = df_merged.groupby('year_month')['total_price'].sum().reset_index()
sales_per_month['year_month'] = sales_per_month['year_month'].astype(str)

plt.figure(figsize=(6, 4))
sns.lineplot(data=sales_per_month, x='year_month', y='total_price', marker='o')

plt.title("Total Sales per Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
